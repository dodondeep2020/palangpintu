```python
import csv
import PySimpleGUI as sg
import cv2
import os
import sys

# counting the numbers
def is_number(s):
    try:
        float(s)
        return True
    except ValueError:
        pass
    try:
        import unicodedata
        unicodedata.numeric(s)
        return True
    except (TypeError, ValueError):
        pass
    return False

# Take image function
def run():
    sg.theme('Dark Grey 14')
    layout = [[sg.Text('NIM:', size =(7, 1), font='Helvetica 14'), sg.InputText('', font='Helvetica 14')],
              [sg.Text('Name:', size =(7, 1), font='Helvetica 14'), sg.InputText('', font='Helvetica 14')],
              [sg.Button('Save', button_color=('white', '#303030'), font='Helvetica 14', size=(20,1)), 
              sg.Button('Cancel', button_color=('white', '#f44336'), font='Helvetica 14', size=(20,1))]]
    window = sg.Window('Add Student', layout, element_justification='c')
    while True:
        event, values = window.read()
        if event == sg.WIN_CLOSED or event == 'Cancel':  # if user closes window or clicks cancel
            window.close()
            break
        elif event == 'Save':
            Id = values[0]
            name = values[1]
            window.close()
    if(is_number(Id)):
        cam = cv2.VideoCapture(0)
        detector = cv2.CascadeClassifier(cv2.data.haarcascades + 
        'haarcascade_frontalface_default.xml')
        sampleNum = 0
        layout = [ [sg.Text("Please look at the camera",font='Helvetica 24')],
                   [sg.Text("Photo is being processed: "),sg.ProgressBar(50, orientation='h', size=(20, 20), key='progressbar')],
                   [sg.Image(filename='', key='image')],[sg.Button("Cancel process",size=(40,1))] ]
        window = sg.Window('Taking a photo', layout, auto_size_buttons=False, element_justification='c', location=(350, 75))
        progress_bar = window['progressbar']
        while(True):
            event, values = window.read(timeout=1)
            if event == "Batalkan proses" or event == sg.WIN_CLOSED:
                cam.release()
                cv2.destroyAllWindows()
                window.close()
                break
            ret, img = cam.read()
            gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
            faces = detector.detectMultiScale(gray, 1.3, 5, minSize=(30,30),flags = cv2.CASCADE_SCALE_IMAGE)
            for(x,y,w,h) in faces:
                cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)
                #incrementing sample number
                sampleNum = sampleNum+1
                progress_bar.UpdateBar(int(sampleNum))
                #saving the captured face in the dataset folder
                cv2.imwrite("data" + os.sep+ "student" + os.sep+ "training_images" + os.sep +name + "."+Id + '.' +
                            str(sampleNum) + ".jpg", gray[y:y+h, x:x+w])
            imgbytes = cv2.imencode(".png", img)[1].tobytes()
            window["image"].update(data=imgbytes)
            #wait for 100 miliseconds
            if cv2.waitKey(100) & 0xFF == ord('q'):
                break
            # break if the sample number is more than 50
            elif sampleNum > 49:
                break
        window.close()
        cam.release()
        cv2.destroyAllWindows()
        # res = "Photo with ID : " + Id +" dan "+ " Name : " + name + " berhasil di simpan"
        row = [Id, name]
        with open("data"+ os.sep + "student"+ os.sep + "mahasiswa.csv", 'a+') as csvFile:
            writer = csv.writer(csvFile)
            writer.writerow(row)
        csvFile.close()
    else:
        run()


import csv
import PySimpleGUI as sg
import cv2
import os
import sys

# counting the numbers
def is_number(s):
    try:
        float(s)
        return True
    except ValueError:
        pass
    try:
        import unicodedata
        unicodedata.numeric(s)
        return True
    except (TypeError, ValueError):
        pass
    return False

# Take image function
def run():
    sg.theme('Dark Grey 14')
    layout = [[sg.Text('NIM:', size =(7, 1), font='Helvetica 14'), sg.InputText('', font='Helvetica 14')],
              [sg.Text('Name:', size =(7, 1), font='Helvetica 14'), sg.InputText('', font='Helvetica 14')],
              [sg.Button('Save', button_color=('white', '#303030'), font='Helvetica 14', size=(20,1)), 
              sg.Button('Cancel', button_color=('white', '#f44336'), font='Helvetica 14', size=(20,1))]]
    window = sg.Window('Add student', layout, element_justification='c')
    while True:
        event, values = window.read()
        if event == sg.WIN_CLOSED or event == 'Cancel':  # if user closes window or clicks cancel
            window.close()
            break
        elif event == 'Save':
            Id = values[0]
            name = values[1]
            window.close()
    if(is_number(Id)):
        cam = cv2.VideoCapture(0)
        detector = cv2.CascadeClassifier(cv2.data.haarcascades + 
        'haarcascade_frontalface_default.xml')
        sampleNum = 0
        layout = [ [sg.Text("Please look at the camera",font='Helvetica 24')],
                   [sg.Text("Photos are being processed: "),sg.ProgressBar(50, orientation='h', size=(20, 20), key='progressbar')],
                   [sg.Image(filename='', key='image')],[sg.Button("Cancel process",size=(40,1))] ]
        window = sg.Window('Taking a photo', layout, auto_size_buttons=False, element_justification='c', location=(350, 75))
        progress_bar = window['progressbar']
        while(True):
            event, values = window.read(timeout=1)
            if event == "Cancel process" or event == sg.WIN_CLOSED:
                cam.release()
                cv2.destroyAllWindows()
                window.close()
                break
            ret, img = cam.read()
            gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
            faces = detector.detectMultiScale(gray, 1.3, 5, minSize=(30,30),flags = cv2.CASCADE_SCALE_IMAGE)
            for(x,y,w,h) in faces:
                cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)
                #incrementing sample number
                sampleNum = sampleNum+1
                progress_bar.UpdateBar(int(sampleNum))
                #saving the captured face in the dataset folder
                cv2.imwrite("data" + os.sep+ "student" + os.sep+ "training_images" + os.sep +name + "."+Id + '.' +
                            str(sampleNum) + ".jpg", gray[y:y+h, x:x+w])
            imgbytes = cv2.imencode(".png", img)[1].tobytes()
            window["image"].update(data=imgbytes)
            #wait for 100 miliseconds
            if cv2.waitKey(100) & 0xFF == ord('q'):
                break
            # break if the sample number is more than 50
            elif sampleNum > 49:
                break
        window.close()
        cam.release()
        cv2.destroyAllWindows()
        # res = "Foto dengan id : " + Id +" dan "+ " Name : " + name + " saved successfully"
        row = [Id, name]
        with open("data"+ os.sep + "student"+ os.sep + "mahasiswa.csv", 'a+') as csvFile:
            writer = csv.writer(csvFile)
            writer.writerow(row)
        csvFile.close()
    else:
        run()
```
